<b>4. Variables & Naming Conventions </b>

<u>Variable</u>: A name that represnets a value stored in the computer memory.

<u>Assignment Statement</u>: Used to create a variable and assign it data

Rules for Naming Variables in Python:
- No key words <a href='https://www.w3schools.com/python/python_ref_keywords.asp'>Key Words in Python</a>
- No spaces
- Must start with a letter or underscore ie cannot start with a digit(but could end with one)
- After first letter, you can use letters, digits or underscores
- Variable names are case sensetive
- Variable Names should reflect its use

#### Variable Examples

```python
num = 5   #integer 
my_str = 'Hello'
_str = 'starts with _ '
num2 = 2.0 #float

print(num)
print(my_str, '\n',_str)

#sum of two numbers
print(num+num2)
```

#### 12. Formatting Numbers
using the built-in function <u>format</u> which takes two arguments(Number to be formatted, Specificed format)

##### Example

```python
# the format specifies that 10 spaces should be there,so the number looks padded
print(format(123456,'10d')) 

#float with 3 decimal spaces after the .
print(format(123456,'0.3f'))

#show percentage.
print(format(0.2,'%'))
```
